```python
import urllib2
from bs4 import BeautifulSoup as bs
#count=0
import pandas as pd
import re
l={}
m=0
for i in range(0,208):
    data=urllib2.urlopen("http://www.seas.upenn.edu/directory/profile.php?ID="+str(i+1))
    soup = bs(data)
    
    if soup.h1 and soup.h2:
        m = m+1
        #count=count+1
        #print count,i+1
        name= soup.h1.contents
        designation=soup.h2.contents
        ######print name, designation,
        name= name[0].encode("ascii", "ignore")
        d=designation
        d1=d[2::2]
        d2=d1[0].encode("latin-1")
        d2=d2.strip()
        #for i in d1:
        #   d2+=str(i.encode("latin-1"))+" "
        #   d2=d2.strip()

        ########print d2
        
        a=soup.find_all("p")
        for i in a:
            if "<strong>Education:</strong>" in str(i):
                #print i.contents
                for j in i.contents:
                    if "phd" in str(j).lower() or "ed.d" in str(j).lower() or "ph.d." in str(j).lower() or "j.d." in str(j).lower():
                        ######print j
                        year = re.findall(r'\d{1,4}', j)
                        if len(year)>0:
                            year = year[0]
                        ######print year
                        break
        l[m]=[name, j,d2,year]
df = pd.DataFrame(l)
df1 = df.T
df1.columns = ['name','Education' , 'Department', 'year']
df1
```
